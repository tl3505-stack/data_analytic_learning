## 1. Non-Linear Regression

### Polynomial Regression (StatsModels)
```python
from statsmodels.formula.api import ols
# MUST include all lower-degree terms!
model = ols('likes ~ time_since + I(time_since**2)', data=train).fit()
print(model.rsquared)
print(model.pvalues['I(time_since ** 2)'])
```

### Step Functions
```python
train['time_bin'] = pd.cut(train.time_since,
    bins=[0, 10000, 30000, 60000, float('inf')],
    labels=['new','mid','old','very_old'])
model_step = ols('likes ~ time_bin', data=train).fit()
# Coefficients = difference from reference (first label)
```

## 2. Feature Selection

### Correlation Filter
```python
from scipy.stats import pearsonr
for col in X_train.columns:
    r, p = pearsonr(X_train[col], y_train)
    print(f'{col}: r={r:.4f}, p={p:.4e}')
```

### VIF (Multicollinearity)
```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
X_c = sm.add_constant(X_train)
vif_data = pd.DataFrame({
    'feature': X_c.columns,
    'VIF': [variance_inflation_factor(X_c.values, i) for i in range(X_c.shape[1])]
})
# VIF > 5 → potential multicollinearity concern
# VIF > 10 → serious problem
```

### Sequential Feature Selection (mlxtend)
```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
sfs = SFS(LinearRegression(), k_features='best', forward=True,
          scoring='r2', cv=5)
sfs.fit(X_train, y_train)
print(sfs.k_feature_names_)
# forward=True → Forward; forward=False → Backward
# floating=True → Stepwise
```

### Lasso (Automatic Shrinkage)
```python
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

### PCA (Dimension Reduction)
```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X_tr_sc)
cum_var = np.cumsum(pca.explained_variance_ratio_)
n_components = np.argmax(cum_var >= 0.60) + 1  # for 60% variance
# Then refit with PCA(n_components=n_components)
pca_k = PCA(n_components=n_components)
```
