## ⚠️ Exam Traps Summary

| Trap | What to remember |
|---|---|
| Polynomial lower-degree terms | `I(x**2)` requires `x` also in formula |
| Step function interpretation | Coef = difference from REFERENCE (first bin) |
| PolynomialFeatures count | 2 vars, degree 2 → **5** features |
| StandardScaler | Fit on TRAIN only, transform both |
| Test set for model selection | Always pick by TEST error, not train |
| VIF threshold | > 5 concern, > 10 serious |
| OOB score | RF's built-in validation score (no separate test set needed) |
| SVM no probabilities | Use `decision_function()` not `predict_proba()` |
| High gamma → overfitting | Train accuracy near 1, test drops |
| Maximal tree → overfitting | Train RMSE = 0 or near-0, test RMSE large |
| GridSearchCV scoring for RMSE | Use `'neg_root_mean_squared_error'` |
| Pseudo R² (logit) | Use `model.prsquared`, not `rsquared` |
| Sensitivity vs Precision | Sensitivity = recall = TP/(TP+FN); FN costly |
| AUC | Use `roc_auc_score(y_true, predict_proba[:,1])` |
