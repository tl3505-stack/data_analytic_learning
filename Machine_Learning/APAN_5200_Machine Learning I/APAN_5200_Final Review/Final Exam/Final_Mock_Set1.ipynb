## 1. Logistic Regression

### Core Formula
```
log(p / (1-p)) = β₀ + β₁x₁ + β₂x₂ + ... + βₚxₚ
p = e^(linear combination) / (1 + e^(linear combination))
```
- Output is **probability** of the event (Y=1), always between 0 and 1
- Estimated by **Maximum Likelihood** (not OLS)

### Key Code — StatsModels
```python
from statsmodels.formula.api import logit
model = logit('sold ~ startprice', data=train).fit()
print(model.summary())          # Pseudo R², p-values
model.predict(train)            # predicted probabilities
model.predict(pd.DataFrame({'startprice': [300]}))  # specific value
# Percent change in odds:
(np.exp(model.params['startprice']) - 1) * 100
```

### Key Code — Sklearn
```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    confusion_matrix, classification_report,
    roc_auc_score, recall_score, precision_score
)
lr = LogisticRegression(max_iter=10000, random_state=617)
lr.fit(X_train, y_train)
pred_train = lr.predict(X_train)        # class label (0 or 1)
proba_train = lr.predict_proba(X_train)[:, 1]  # probability of class 1

# Metrics:
auc = roc_auc_score(y_test, lr.predict_proba(X_test)[:, 1])
sensitivity = recall_score(y_test, pred_test)   # TP / (TP + FN)
precision = precision_score(y_test, pred_test)  # TP / (TP + FP)
```

### Coefficient Interpretation (⚠️ EXAM TRAP)
- Coefficient in logit → change in **log-odds**
- `exp(coef)` → **odds ratio** (multiplicative change in odds)
- `(exp(coef) - 1) * 100` → **% change in odds**
- Positive coef → increases probability; Negative → decreases

### Metrics Quick Reference
| Metric | Formula | Use when... |
|---|---|---|
| **Sensitivity/Recall** | TP/(TP+FN) | FN is costly (e.g., missing sick patients) |
| **Precision** | TP/(TP+FP) | FP is costly (e.g., false alarms) |
| **Specificity** | TN/(TN+FP) | FP is costly |
| **Accuracy** | (TP+TN)/N | Overall correctness |
| **AUC** | 0.5 (random) to 1.0 | Overall discrimination ability |

## 4. Support Vector Machines

### Pipeline (ALWAYS scale first!)
```python

from sklearn.svm import SVC, LinearSVC

scaler = StandardScaler()

X_te = scaler.transform(X_te_raw)      # transform test

# Linear:

# Non-linear (RBF):

```

### SVM vs Logistic Regression
| | SVM | Logistic Reg |
|---|---|---|
| Probabilities | ❌ No | ✅ Yes |
| Interpretable | ❌ No p-values | ✅ Coefficients |
| Non-linear | ✅ Kernel trick | ❌ Manual features |
