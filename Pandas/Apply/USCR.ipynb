```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv"
crime = pd.read_csv(url, sep=",")
crime.head(10)
```

```python
print(type(crime.shape[1]))
```

```python
crime.info()
```

```python
# Step 5. Convert the type of the column Year to datetime64

crime['Year'] = pd.to_datetime(crime['Year'], format='%Y', errors='coerce')
print(crime['Year'].info())
```

```python
# Step 6. Set the Year column as the index of the dataframe
crime.set_index('Year')
```

```python
crime.head()
```

```python
# Step 7. Delete the Total column
crime = crime.drop('Total', axis=1)
crime.head()
```

```python
# Step 8. Group the year by decades and sum the values
# Pay attention to the Population column number, summing this column is a mistake

crime_decades = crime.resample('10AS').sum(numeric_only=True)
population = crime['Population'].resample('10AS').max()
crime_decades['Population'] = population
crime_decades
```

```python
# Step 9. What is the most dangerous decade to live in the US?
MDD = crime_decades['Murder'].idxmax()
print(MDD)
```

```python
# 创造一个新指标：犯罪率 = 暴力犯罪总数 / 总人口
crime_decades['Violent_Rate'] = crime_decades['Violent'] / crime_decades['Population']
crime_decades['Violent_Rate'].idxmax()
```
