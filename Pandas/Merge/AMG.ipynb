```python
import pandas as pd
```

```python
url1 = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv'
url2 = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv'

cars1 = pd.read_csv(url1, sep=",")
cars2 = pd.read_csv(url2, sep=",")

cars1.head()
```

```python
cars2.head()
```

```python
# Step 4. Oops, it seems our first dataset has some unnamed blank columns, fix cars1
# First methond, drop the columns by hands
cars1.drop(['Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11',	'Unnamed: 12',	'Unnamed: 13'], axis = 1)
```

```python
# second methond, auto drop
cars1 = cars1.loc[:, ~cars1.columns.str.contains('Unnamed')]
print(cars1)
```

```python
# method 3, 创建无 NaN 新列表”
cars1.dropna(axis=1, how='all')
```

```python
# method 4, 正面选择 (Selection)
cars1 =cars1.loc[:, "mpg":"car"]
print(cars1)
```

```python
# Step 5. What is the number of observations in each dataset?
print(cars1.shape[0])
print(cars2.shape[0])
```

```python
cars = pd.concat([cars1, cars2], axis=0, ignore_index=True)
cars.tail()
```

```python
# Step 7. Oops, there is a column missing, called owners. Create a random number Series from 15,000 to 73,000.
import numpy as np

random_owners = np.random.randint(15000, 73000, size=len(cars))

cars['owners'] = random_owners

print(cars)
```
